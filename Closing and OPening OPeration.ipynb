## Closing and OPening OPeration

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load Image
image = cv2.imread("beauty.jpg", 0)

# Kernel
kernel = np.ones((5, 5), np.uint8)

# Closing Operation
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# Opening Operation
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Display Images
plt.figure(figsize=(12, 8))

plt.subplot(131), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.axis('off')

plt.subplot(132), plt.imshow(closing, cmap='gray')
plt.title('Closing Operation'), plt.axis('off')

plt.subplot(133), plt.imshow(opening, cmap='gray')
plt.title('Opening Operation'), plt.axis('off')

plt.show()
```
